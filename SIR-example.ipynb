### Import the SIR model
Import the model and specify the following parameters:
- N is the total population
- beta is related to the virus spread velocity
- gamma is related to the average time to recover from the illness
- I0 is the number of initial infected people
- R0 is the number of initial recovered people

```python
from src.simple_models import SIR

time, susceptible, infected, removed = SIR(N = 10**6,
                                           beta = 0.4, 
                                           gamma = 0.1,
                                           I0=1,
                                           R0=0)
```
